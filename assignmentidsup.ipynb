```python
#we are importing 1000 ages randomly
import random

Age = []
for i in range(1000):
    age = random.randint(60,70)
    Age.append(age);
Age
```

```python
#Mean and standard deviation of age
import numpy as np

Age = np.array(Age)


mean = np.mean(age)
std= np.std(Age)
mean,std
```

```python
#plotting of graph using histogram 
import matplotlib.pyplot as plt

plt.hist(Age,bins= 22)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Frequency chart of ages")
plt.show()
```

```python
'''In this case, since we 
generated the ages randomly from a uniform
distribution between 60 and 70, the
distribution of ages should be
approximately uniform, and thus not normal.'''
```

```python
#choosing randomly 30 numbers from age
sample= np.random.choice(Age,size = 30)
```

```python
#mean of sample
s_mean = []
s_mean.append(np.mean(sample))
s_mean
```

```python
#repetition of step 4 and 5 for 500 times
for i in range(0,499):
    sample= np.random.choice(Age,size = 30)
    s_mean.append(np.mean(sample))
```

```python
np.size(s_mean)
```

```python
plt.hist(s_mean,bins=20)
plt.xlabel("sample mean")
plt.ylabel("frequency")
plt.title("Frequecy vs sample_mean")
```

```python
#mean and standard deviation
s_mean = np.array(s_mean)
np.mean(s_mean),np.std(s_mean)
```

```python
np.mean(s_mean),mean
```

```python
np.std(s_mean),std/(30**0.5)
```
